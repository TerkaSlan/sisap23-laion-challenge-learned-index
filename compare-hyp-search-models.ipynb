```python
import os
base_path = '/auto/brno12-cerit/nfs4/home/tslaninakova/sisap-challenge/models'
csvs = [f for f in os.listdir(base_path) if f.endswith('.csv')]
```

```python
import pandas as pd
import numpy as np
df = pd.DataFrame(csvs, columns=['model'])
```

```python
df['avg_recall'] = df.model.apply(lambda x: pd.read_csv(f'{base_path}/{x}')['anns_found'].mean() if len(pd.read_csv(f'{base_path}/{x}').columns) > 2 else np.nan)
df['avg_navigation_time'] = df.model.apply(lambda x: pd.read_csv(f'{base_path}/{x}')['navigation_t'].mean() if len(pd.read_csv(f'{base_path}/{x}').columns) > 2 else np.nan)
df['avg_search_time'] = df.model.apply(lambda x: pd.read_csv(f'{base_path}/{x}')['search_t'].mean() if len(pd.read_csv(f'{base_path}/{x}').columns) > 2 else np.nan)
```
